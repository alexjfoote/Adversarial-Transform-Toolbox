```python
from attacks import PGD
from metric import Metric
from sklearn.metrics import accuracy_score

parameters = {
                "epsilon": 0.1,
                "steps": 10,
                "norm": "l2",
                "C": 2,
                "margin": 1,
                "input_range": (0, 255)
            }

Pgd = PGD(parameters)

def compute_accuracy(outputs, labels):
    predictions = output.argmax(dim=1, keepdim=False)
    return accuracy_score(labels, predictions)

Accuracy = Metric(compute_accuracy, Pgd, model, test_data)
```

```python
acc = Accuracy.compute_metric(adversarial=False)
robust_acc = Accuracy.compute_metric(adversarial=True, n=200)
print(f"Accuracy: {acc}, robust accuracy: {robust_acc}")
```
